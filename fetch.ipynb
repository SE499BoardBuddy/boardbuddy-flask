```python
import requests
import pandas as pd
import time
import xml.etree.ElementTree as ET

read_df = pd.read_csv('boardgames_ranks.csv', nrows=1000)
i = 0
to_keep_list = ['yearpublished', 'minplayers', 'maxplayers', 'playingtime',
                'minplaytime', 'maxplaytime', 'age', 'name', 'description', 'thumbnail',
                'image', 'boardgamepublisher',
                'boardgamecategory',
                'videogamebg', 'boardgamedesigner', 'boardgameartist',
                'boardgamemechanic',
                'boardgamesubdomain']
column_list = ['id','bayes_average','year_published', 'min_players', 'max_players', 'playing_time',
                'min_playtime', 'max_playtime', 'age', 'name', 'description', 'thumbnail',
                'image', 'boardgame_publisher',
                'boardgame_category',
                'videogame_bg', 'boardgame_designer', 'boardgame_artist',
                'boardgame_mechanic',
                'boardgame_subdomain']
df = pd.DataFrame()

while i <= 999:
    list_of_num = ''
    for num in read_df['id'].loc[i:i+9]:
        if len(list_of_num) == 0:
            list_of_num = str(num)
        else:
            list_of_num = list_of_num + ','  + str(num)
        i += 1

    print(list_of_num)
    response = requests.get("https://api.geekdo.com/xmlapi/boardgame/"+str(list_of_num))
    print(response.status_code)

    root = ET.fromstring(response.content)
    queues = root.findall('./boardgame')
    games = []
    for queue in queues:
        game_info = [queue.attrib.get('objectid'), read_df.loc[read_df['id'] == int(queue.attrib.get('objectid'))]['bayesaverage'].iloc[0]]
        for j in to_keep_list:
            if(j == 'name'):
                for name_tag in queue.findall('name'):
                    if(name_tag.attrib.get('primary') != None and name_tag.attrib.get('primary') == 'true'):
                        game_info.append(name_tag.text)
            else:
                tag = queue.find(j)
                game_info.append(tag.text if tag != None else '')
        games.append(game_info)
    # print(games)
    tempDf = pd.DataFrame(games, columns=column_list)
    df = pd.concat([df, tempDf])
    print(i, len(df.index))
    time.sleep(10)

df.to_csv('bgg_games_info.csv', index=False)
print("fetch complete")
```
